```python
# show image stats

print('Image size as array: %d x %d' % (camera.shape[0], camera.shape[1]))
print('Pixel number of image: %d' % camera.size)
print('Pixel range: [%d, %d]' % (camera.min(), camera.max()))
print('Average pixel value: %.2f' % camera.mean())
```

```python
# load custom image source

from skimage import io
colony = io.imread('yeast_colony_array.jpg')
print(type(colony))

# number array dimensions and position
print(colony.shape)
```

```python
# position figure images side by side
plt.figure(figsize=(24, 18))

# 01 - all channels real shape plot
plt.subplot(121)
# plt.imshow(colony)          # plot numpy arrays as images
plt.imshow(colony[:, :, :])   # apply pseudocolor schemes to image plots (img[R, G, B])
plt.title('3-Channel Image')
plt.axis('on')

# 02 - one channel only plot
plt.subplot(122)
plt.imshow(colony[:, :, 0])
plt.title('1-Channel Image')
plt.axis('on')
```

```python
# apply pseudocolor schemes to image plots

colony_clr = colony[:, :, 0]
plt.imshow(colony_clr)
```

```python
# 2D, no color = a luminosity image
# viridis = default colormap (aka lookup table, LUT)
# apply LUT 

plt.imshow(colony_clr, cmap = 'hot') 
```

```python
# change colormaps on existing plot objects

# colony_clr = colony[:, :, 0]
# plt.imshow(colony_clr, cmap = 'hot')

colony_cmap = plt.imshow(colony_clr)
colony_cmap.set_cmap('nipy_spectral')
```
